## Statistics

```python
df_finance['date'] = pd.to_datetime(df_finance['date'])
df_sport['date'] = pd.to_datetime(df_sport['date'])
df_election['date'] = pd.to_datetime(df_election['date'])
```

```python
df_finance['url'].value_counts()
```

```python
df_finance['word_count'] = df_finance['summary'].str.split().str.len()
df_finance['flesch_kincaid'] = df_finance['summary'].apply(textstat.flesch_reading_ease)

df_sport['word_count'] = df_sport['summary'].str.split().str.len()
df_sport['flesch_kincaid'] = df_sport['summary'].apply(textstat.flesch_reading_ease)

df_election['word_count'] = df_election['summary'].str.split().str.len()
df_election['flesch_kincaid'] = df_election['summary'].apply(textstat.flesch_reading_ease)

df_general['word_count'] = df_general['summary'].str.split().str.len()
df_general['flesch_kincaid'] = df_general['summary'].apply(textstat.flesch_reading_ease)
```

```python
df_general[['word_count', 'flesch_kincaid']].describe()
```
