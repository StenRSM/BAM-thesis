## Only keep unique articles

```python
df_finance
```

```python
df_finance_unique = df_finance.drop_duplicates(subset="title", keep="first").drop_duplicates(subset='summary', keep='first').reset_index()
df_sport_unique = df_sport.drop_duplicates(subset="title", keep="first").drop_duplicates(subset='summary', keep='first').reset_index()
df_election_unique = df_election.drop_duplicates(subset="title", keep="first").drop_duplicates(subset='summary', keep='first').reset_index()
```

## String operations

```python
df_sport_unique
```

```python
df_finance_unique['summary'] = df_finance_unique['summary'].apply(lambda x: x.replace('\n', ' '))
df_sport_unique['summary'] = df_sport_unique['summary'].apply(lambda x: x.replace('\n', ' '))
df_election_unique['summary'] = df_election_unique['summary'].apply(lambda x: x.replace('\n', ' '))
```

```python
df_finance_unique['summary'] = df_finance_unique['summary'].apply(lambda x: x.replace('\xa0', ' '))
df_sport_unique['summary'] = df_sport_unique['summary'].apply(lambda x: x.replace('\xa0', ' '))
df_election_unique['summary'] = df_election_unique['summary'].apply(lambda x: x.replace('\xa0', ' '))
```

## Write data as new file

```python
df_finance_unique.to_csv(r'C:\Users\Sten\Documents\EUR BAM\Thesis\data\news\finance_processed.csv', index=False)
df_sport_unique.to_csv(r'C:\Users\Sten\Documents\EUR BAM\Thesis\data\news\sport_processed.csv', index=False)
df_election_unique.to_csv(r'C:\Users\Sten\Documents\EUR BAM\Thesis\data\news\election_processed.csv', index=False)
```
